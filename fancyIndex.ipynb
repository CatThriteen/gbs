```python
@jit(nopython=True, fastmath=True)
def fancyIndex(x, idx):
    res = np.empty((len(idx),len(x),idx.shape[1]), dtype=x.dtype)
    for i in range(len(idx)):
        res[i] = x[:,idx[i]]
    return res
```
